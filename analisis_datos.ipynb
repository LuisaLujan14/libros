```python
# Eliminar el guion '-' en el campo ISBN (978-0451524935) del registro 3 de la data, debido a que el identificador del libro es numerico 
df['isbn'] = df['isbn'].str.replace('-', '')
df.head(10)

```

```python
# Mapeo manual de ubicaciones a coordenadas
coordenadas = {
    'España': (40.4168, -3.7038),  # Coordenadas de España
    'Colombia': (4.5709, -74.2973),  # Coordenadas de Colombia
    'Reino Unido': (51.5074, -0.1278),  # Coordenadas de Reino Unido
    'Estados Unidos': (37.0902, -95.7129),  # Coordenadas de Estados Unidos
    'Rusia': (61.5240, 105.3188)  # Coordenadas de Rusia
}

# Agregar columnas de coordenadas al DataFrame
df['latitud'] = df['ubicacion_editorial'].map(lambda x: coordenadas[x][0] if x in coordenadas else None)
df['longitud'] = df['ubicacion_editorial'].map(lambda x: coordenadas[x][1] if x in coordenadas else None)

# Ver el Data resultante actualizado
df.head(10)
```
